```python
%pip install pandas matplotlib seaborn plotly dash scikit-learn joblib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import seaborn as sns
from datetime import datetime
```

```python
accidents = pd.read_csv("cleaned_data/merged.csv")
path = "report/merged/"
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=accidents, x='worker_experience_in_years', y='program',
    alpha=0.6, color='teal'
)
plt.title("Worker Experience vs Program", fontsize=14)
plt.xlabel("Worker Experience (Years)", fontsize=12)
plt.ylabel("Program Categories", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# plt.tight_layout()
plt.savefig(path + "worker_experience_vs_program.png")
plt.show()
plt.close()
```
