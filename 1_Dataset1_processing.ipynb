```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'cleaned_data/events.csv'
dataset = pd.read_csv(file_path)
dataset.columns = dataset.columns.str.strip()
dataset['EVENT DATE'] = pd.to_datetime(dataset['date'], errors='coerce', dayfirst=True)
# dataset_cleaned = dataset.dropna(subset=['EVENT DATE', 'sector_description', 'region'])

# Monthly Distribution of Workplace Events
dataset['EVENT MONTH'] = dataset['EVENT DATE'].dt.to_period('M')
events_per_month = dataset['EVENT MONTH'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(events_per_month.index.astype(str), events_per_month.values, marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Distribution of Workplace Events')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.grid(True)
plt.tight_layout()
plt.savefig("report/dataset1/trend.png")
plt.show()
```

```python
# Sector-wise Distribution
sector_counts = dataset['sector_description'].value_counts().head(30)

plt.figure(figsize=(12, 6))
sector_counts.plot(kind='bar')
plt.title('Sector-wise Distribution of Workplace Events')
plt.xlabel('Sector Description')
plt.ylabel('Number of Events')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("report/dataset1/sectors.png")
plt.show()
```

```python

# Regional Distribution
region_counts = dataset['city'].value_counts().head(20)

plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='skyblue')
plt.title('Regional Distribution of Workplace Events')
plt.xlabel('Municipality')
plt.ylabel('Number of Events')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("report/dataset1/cities.png")
plt.show()
```

```python

# Workforce Size Distribution
worker_count_distribution = dataset['worker_count'].value_counts()

plt.figure(figsize=(8, 5))
worker_count_distribution.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Workplace Events by Workforce Size')
plt.xlabel('Workforce Size')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("report/dataset1/worker_count.png")
plt.show()
```
